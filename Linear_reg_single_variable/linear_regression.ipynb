# **Linear Regression with Single Variable.**

#### Example 1: Predicting income from years(canada capita income )

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('canada_per_capita_income.csv')
df
```

```python
plt.scatter(df.year,df.income,color='red',marker='+')
```

##### Spliting the model into training and testing data

```python
X = df[['year']]
y = df.income
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size = 0.2)
```

```python
regression = linear_model.LinearRegression()
regression.fit(X_train,y_train) # Training the model
```

```python
regression.predict(X_test) # Predicting the values based on the test data
```

```python
y_test
```

```python
regression.score(X_test,y_test)*100
```

```python
plt.plot(df.year,regression.predict(df[['year']]),color='blue')
plt.scatter(df.year,df.income,color='red',marker='+')
plt.xlabel('Year')
plt.ylabel('Income')
plt.title('Canada Income Prediction')
plt.show()
```

#### Example 2 : Predicting Car price from mileage

```python
df1 = pd.read_csv('car_price_prediction.csv')
df1
```

```python
df1.isnull().sum()
```

```python
plt.scatter(df1.Cylinders,df1.Price,color='red',marker='+')
plt.xlabel('Cylinders')
plt.ylabel('Price')
plt.title('Car Price Prediction')
plt.show()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df1[['Cylinders']],df1.Price)
```

```python
reg.predict(np.array(2).reshape(-1,1))
```

```python
plt.scatter(df1.Cylinders,df1.Price,color='red',marker='+')
plt.plot(df1.Cylinders,reg.predict(df1[['Cylinders']]),color='blue')
plt.xlabel('Cylinders')
plt.ylabel('Price')
plt.title('Car Price Prediction')
plt.show()
```
